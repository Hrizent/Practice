# Задание 1

Задание:

Создайте класс с именем Rectangle который имеет:
- Атрибуты ширины и высоты.
- Метод расчета площади.
- Метод расчета периметра.
- Метод отображения размеров прямоугольника.

Создайте экземпляр класса Rectangleи продемонстрируйте его функциональность.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width=width
        self.height=height


    def calc_area(self):
        return f'Прямоугольник имеет площадь {self.width*self.height}'


    def calc_perimeter(self):
        return f'Прямоугольник имеет периметер {2*(self.width+self.height)}'


    def show_size(self):
        return f'Прямоугольник имеет ширину {self.width} и высоту {self.height}'


rect=Rectangle(2,5)
print(rect.calc_area())
print(rect.calc_perimeter())
print(rect.show_size())
```

Задание: Создайте мини версию банковской системы:


Инструкции:

1. Создайте класс BankAccountсо следующими атрибутами:
    - account_holder -  владелец счета
    - balance - баланс счета

2. Реализуйте следующие методы:
    - Метод для инициализации владельца счета: имя владельца счета и установите начальный баланс на 0.
    - deposit(amount): Добавьте указанную сумму к балансу.
    - withdraw(amount): Вычесть указанную сумму из баланса, если средств достаточно; в противном случае вывести предупреждение.
    - get_balance(): Возврат текущего баланса.


Создайте объект класса и продемонстрируйте его возможности

```python
class BankAccount:
    def __init__(self, account_holder=0, balance=0):
        self.account_holder = account_holder
        self.balance = balance


    def deposit(self, amount):
        self.balance+=amount
        return f'сумма {amount} успешно внесена на счет'


    def withdraw(self, amount):
        if self.balance>=amount:
            self.balance-=amount
            return f'сумма {amount} успешно внесена на счет'
        else:
            return 'недостаточно средств'


    def get_balance(self):
        return f'на счёте {self.account_holder} {self.balance} средств'


bank_acc = BankAccount('tsaregorodtsev', 1000)
print(bank_acc.deposit(256))
print(bank_acc.withdraw(6))
print(bank_acc.get_balance())
print(bank_acc.withdraw(2000241))
```

# Задание 3

Возьмите код и задание (Рыцарь и дракон) из предыдущей практики и реализуйте его с применением классов

```python
from random import randint

class Knight:
    def __init__(self):
        name = input("введите рыцаря имя ")
        armor = int(input("введите значение защиты брони "))
        weapon_attack, weapon = input("введите значение атаки оружия рыцаря и что это за оружие через запятую ").split(', ')
        health = int(input("введите значение здоровья рыцаря "))
        self.name = name
        self.armor = armor
        self.weapon_attack = int(weapon_attack)
        self.weapon = weapon
        self.health = health*(1+armor/100)


    def attack(self):
        return self.weapon_attack+randint(-2, 2)


    def take_damage(self, dragon_damage):
        self.health -= dragon_damage
        return f'Дракон атакует, здоровье рыцаря стало {self.health}!\n'


    def heal(self):
        self.heal_tries = 3
        if self.heal_tries > 0:
            self.health += 20
            self.heal_tries -= 1
            return f'вы полечились на 20 единиц здоровья\n. Ваше здоровье {self.health}'
        else:
            return 'закончились хилки\n'


    def get_info(self):
        return f'здоровье рыцаря {self.name} - {self.health}'


class Dragon:
    def __init__(self):
        name = input("введите имя дракона ")
        health = int(input("введите значение здоровья дракона "))
        damage = int(input("введите значение урона дракона "))
        self.name = name
        self.weapon_attack = damage
        self.health = health


    def attack(self):
        return self.weapon_attack+randint(-2, 2)


    def take_damage(self, knight_damage):
        self.health -= knight_damage
        return f'Рыцарь атакует, здоровье дракона стало {self.health}!\n'


    def get_info(self):
        return f'здоровье дракона {self.name} - {self.health}\n'


class Game:
    def __init__(self):
        self.knight = Knight()
        self.dragon = Dragon()
        self.intro = '''когда то давным давно жил дракон, который не давал покоя деревушке неподалёку от его логова.
но в один день родился мальчик, которому предстояло сразиться с драконом. судьба наделила его огромной физической силой, а
отец воспитал в нем настоящего воина. и вот рыцарь и дракон встречаются в жестокой битве дальше хз'''


    def start_game(self):
        print(self.intro)
        print('\nвы ходите первым\n')
        while True:
            action = input('что хотите сделать? 0 - атаковать, 1 - полекаться ')
            if action == '0':
                print(self.dragon.take_damage(self.knight.attack()))
            elif action == '1':
                print(self.knight.heal())
            else:
                print('нет команды такой ты шо дурак')
            print(self.knight.take_damage(self.dragon.attack()))
            print(self.dragon.get_info())
            print(self.knight.get_info())
            if self.knight.health<=0:
                print('но даже казалось бы могущественный рыцарь не смог одолеть древнего дракона...END')
                break
            if self.dragon.health <= 0:
                print('дракон побежден! в деревню пришел мир HAPPY END')
                break


game=Game()
game.start_game()
```
